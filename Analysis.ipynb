```python
import pandas as pd
from glob import glob
```

```python
datasets = glob("data/*.csv")

li = []

for ds in datasets:
    df = pd.read_csv(ds, index_col=None, header=0)
    li.append(df)
    
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
def normalize_part(c):
    if not c.isnan():
        return c.strip().lower()
    else:
        return c

df["part_normalized"] = df["part"].apply(normalize_part)
```

```python
total_n = len(df)
```

```python
costs = df[["part", "total_paid"]].groupby(["part"]).mean()
```

```python
counts = df["part"].value_counts()
```

```python
pd.options.display.float_format = '{:.2f} GBP'.format
```

```python
joined = pd.concat([costs, counts], axis=1).rename(columns={"total_paid": "mean_price", "part": "count"})
```
